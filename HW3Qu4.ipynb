```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

In this exercise, we arew given the following:

1. **X** is a **Gaussian Model**, with mean zero and variance one
2. We will be using **two** different noise function: **E1** is also a **Gaussian Model**, with mean zero and variance one and **E2** is a **Uniform Model**, with a range from -1 to 1.
3. We will solve the problems by **Y1** and **Y2**
4. **Y1** is the following equation: $Y1=aX+E1$, where we assume that $a=0.5$
5. **Y2** is the following equation: $Y2=aX+E2$, where we assume that $a=0.5$

```python
X=np.random.normal(0,1,1000)
E1=np.random.normal(0,1,1000)
E2=np.random.uniform(-1,1,1000)
```

```python
a=0.5

Y1=a*X+E1
Y2=a*X+E2
```

```python
fontsize=15
fig, ax = plt.subplots(2,2, figsize=(10,10))

ax[0,0].scatter(X, E1)
ax[0,0].set_title("Scatter Plot for X and E1", fontsize=fontsize)
ax[0,0].set_ylabel("E1~N(0,1)")
ax[0,0].set_xlabel("X~N(0,1)")

ax[0,1].scatter(X, Y1)
ax[0,1].set_title("Scatter Plot for X and Y1", fontsize=fontsize)
ax[0,1].set_ylabel("Y1=0.5*X+E1")
ax[0,1].set_xlabel("X~N(0,1)")

ax[1,0].scatter(X, E2)
ax[1,0].set_title("Scatter Plot for X and E2", fontsize=fontsize)
ax[1,0].set_ylabel("E2~Uni(-1,1)")
ax[1,0].set_xlabel("X~N(0,1)")

ax[1,1].scatter(X, Y2)
ax[1,1].set_title("Scatter Plot for X and Y2", fontsize=fontsize)
ax[1,1].set_ylabel("Y2=0.5*X+E2")
ax[1,1].set_xlabel("X~N(0,1)")

#plt.savefig("ABC")
```

```python
x_mle1=(X.T@Y1)/(X.T@X)
error1x=abs(a-x_mle1)/(a)*100

x_mle2=(X.T@Y2)/(X.T@X)
error2x=abs(a-x_mle2)/(a)*100

print("-------------------------------------------------------------------------------------------------------------------------------")
print(f"| > The estimate of the coefficient of X in the estimated regression model for Y1 is: {round(x_mle1, 3)} with a percentage error of {round(error1x, 2)} % |")
print("-------------------------------------------------------------------------------------------------------------------------------") 
print(f"| > The estimate of the coefficient of X in the estimated regression model for Y2 is: {round(x_mle2, 3)} with a percentage error of {round(error2x, 2)} % |")
print("-------------------------------------------------------------------------------------------------------------------------------")


print("   ")    
y_mle1=(X.T@Y1)/(Y1.T@Y1)
error1y=abs(a-y_mle1)/(a)*100

y_mle2=(X.T@Y2)/(Y2.T@Y2)
error2y=abs(a-y_mle2)/(a)*100

print("---------------------------------------------------------------------------------------------------------------------------------")
print(f"| > The estimate of the coefficient of Y1 in the estimated regression model for Y1 is: {round(y_mle1, 3)} with a percentage error of {round(error1y, 2)} % |")
print("---------------------------------------------------------------------------------------------------------------------------------") 
print(f"| > The estimate of the coefficient of Y2 in the estimated regression model for Y2 is: {round(y_mle2, 3)} with a percentage error of {round(error2y, 2)} % |")
print("---------------------------------------------------------------------------------------------------------------------------------")
```

## iv). Is the causal direction between X and Y identifiable from their joint distribution in this case? Briefly explain why.

### Identifiability in Statistical Models
Identifiability is a fundamental concept in statistical modeling, referring to the ability to uniquely determine the model parameters from the probability distribution of the observed data. In simpler terms, a model is identifiable if different values of parameters lead to different distributions of the data.

#### Model Y1: Non-Identifiable Case
- **Definition:** In the context of your first model **(Y1)**, non-identifiability implies that it's impossible to uniquely determine the direction of the causal relationship between variables based on the observed data.
- **Independent Residuals:** The key aspect here is the independence of residuals. In many statistical models, especially linear ones, the independence of residuals is a common assumption. However, if the residuals (errors) are independent of each other and independent of the predictors, it might lead to a situation where multiple sets of parameters can explain the data equally well, hence the non-identifiability.
- **Implications:** This means that, for **Y1**, even with an infinite amount of data, we cannot conclusively determine the direction of causation or the exact parameters of the model.

#### Model Y2: Identifiable Case in LiNGAM
- **Linear Non-Gaussian Models (LiNGAM):** This approach is based on the assumption that the underlying data generation process is linear, but the distributions of the error terms are non-Gaussian. This non-Gaussianity is crucial for identifiability.
- **Identifiability in LiNGAM:** In the case of **Y2**, the non-Gaussian nature of the residuals allows for the identifiability of the model. This is because, under the LiNGAM framework, only one ordering of the variables (one direction of causation) will lead to a fully independent set of residuals.
- **Mathematical Formulation:** Typically, LiNGAM can be represented as  **Y2=BX+ϵ**, where **B** is a matrix of coefficients, **X** is a vector of independent variables, and ϵ is a vector of non-Gaussian residuals. The identifiability comes from the fact that only one specific **B** will render the components of **ϵ** independent.

#### Advanced Considerations
- **Statistical Tests for Non-Gaussianity:** To establish identifiability in LiNGAM, one must employ statistical tests to confirm the non-Gaussian nature of the residuals. Techniques like skewness/kurtosis tests, or more advanced methods like Independent Component Analysis (ICA), can be used.
- **Implications for Causal Inference:** The identifiability in models like **Y2** opens the door for causal inference, which is a significant step beyond mere correlation analysis. This is particularly important in fields like econometrics, psychology, and systems biology.
- **Challenges and Limitations:** It's important to note that while LiNGAM provides a framework for identifiability in certain cases, it comes with its own set of assumptions and limitations. For instance, the linearity assumption might not hold in all real-world scenarios, and the detection of non-Gaussianity can be challenging in practice.

### Conclusion
In summary, the distinction between the models **Y1** and **Y2** in terms of identifiability hinges on the nature of the residuals and the underlying assumptions about the data distribution. While **Y1** suffers from non-identifiability due to independent residuals, **Y2**, under the LiNGAM framework, achieves identifiability through the exploitation of non-Gaussian residuals, enabling more robust causal inference. This discussion is central to advanced statistical modeling and has profound implications in various fields where understanding the direction of causality is crucial.

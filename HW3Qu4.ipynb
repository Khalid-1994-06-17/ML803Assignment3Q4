```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

In this exercise, we arew given the following:

1. **X** is a **Gaussian Model**, with mean zero and variance one
2. We will be using **two** different noise function: **E1** is also a **Gaussian Model**, with mean zero and variance one and **E2** is a **Uniform Model**, with a range from -1 to 1.
3. We will solve the problems by **Y1** and **Y2**
4. **Y1** is the following equation: $Y1=aX+E1$, where we assume that $a=0.5$
5. **Y2** is the following equation: $Y2=aX+E2$, where we assume that $a=0.5$

```python
X=np.random.normal(0,1,1000)
E1=np.random.normal(0,1,1000)
E2=np.random.uniform(-1,1,1000)
```

```python
a=0.5

Y1=a*X+E1
Y2=a*X+E2
```

```python
fontsize=15
fig, ax = plt.subplots(2,2, figsize=(10,10))

ax[0,0].scatter(X, E1)
ax[0,0].set_title("Scatter Plot for X and E1", fontsize=fontsize)
ax[0,0].set_ylabel("E1~N(0,1)")
ax[0,0].set_xlabel("X~N(0,1)")

ax[0,1].scatter(X, Y1)
ax[0,1].set_title("Scatter Plot for X and Y1", fontsize=fontsize)
ax[0,1].set_ylabel("Y1=0.5*X+E1")
ax[0,1].set_xlabel("X~N(0,1)")

ax[1,0].scatter(X, E2)
ax[1,0].set_title("Scatter Plot for X and E2", fontsize=fontsize)
ax[1,0].set_ylabel("E2~Uni(-1,1)")
ax[1,0].set_xlabel("X~N(0,1)")

ax[1,1].scatter(X, Y2)
ax[1,1].set_title("Scatter Plot for X and Y2", fontsize=fontsize)
ax[1,1].set_ylabel("Y2=0.5*X+E2")
ax[1,1].set_xlabel("X~N(0,1)")

#plt.savefig("ABC")
```

```python
x_mle1=(X.T@Y1)/(X.T@X)
error1x=abs(a-x_mle1)/(a)*100

x_mle2=(X.T@Y2)/(X.T@X)
error2x=abs(a-x_mle2)/(a)*100

print("-------------------------------------------------------------------------------------------------------------------------------")
print(f"| > The estimate of the coefficient of X in the estimated regression model for Y1 is: {round(x_mle1, 3)} with a percentage error of {round(error1x, 2)} % |")
print("-------------------------------------------------------------------------------------------------------------------------------") 
print(f"| > The estimate of the coefficient of X in the estimated regression model for Y2 is: {round(x_mle2, 3)} with a percentage error of {round(error2x, 2)} % |")
print("-------------------------------------------------------------------------------------------------------------------------------")


print("   ")    
y_mle1=(X.T@Y1)/(Y1.T@Y1)
error1y=abs(a-y_mle1)/(a)*100

y_mle2=(X.T@Y2)/(Y2.T@Y2)
error2y=abs(a-y_mle2)/(a)*100

print("---------------------------------------------------------------------------------------------------------------------------------")
print(f"| > The estimate of the coefficient of Y1 in the estimated regression model for Y1 is: {round(y_mle1, 3)} with a percentage error of {round(error1y, 2)} % |")
print("---------------------------------------------------------------------------------------------------------------------------------") 
print(f"| > The estimate of the coefficient of Y2 in the estimated regression model for Y2 is: {round(y_mle2, 3)} with a percentage error of {round(error2y, 2)} % |")
print("---------------------------------------------------------------------------------------------------------------------------------")
```

## ii). Perform linear regression of Y on X on the sample. Give the estimate of the coefficient of X in the estimated regression model. Draw the scatter plot of X and the estimated noise term. Do you think they are statistically independent, based on visual inspection?

```python
y1=x_mle1*X
x1=y_mle1*Y1
y2=x_mle2*X
x2=y_mle2*Y2
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X, Y1, color='blue')
plt.title("Scatter Plot for X and Y1", fontsize=fontsize)
plt.plot(X, y1, color='red', linewidth=5)
plt.ylabel("Y1")
plt.xlabel("X")
plt.grid()

plt.figure(figsize=(5,5))
plt.scatter(X, Y2, color='blue')
plt.title("Scatter Plot for X and Y2", fontsize=fontsize)
plt.plot(X, y2, color='red', linewidth=5)
plt.ylabel("Y2")
plt.xlabel("X")
plt.grid()

print(f"> The estimate of the coefficient of X in the estimated regression model Y1 is approximately {round(x_mle1, 3)} and they are statistically independent, based on visual inspection.")  
print(f"> The estimate of the coefficient of X in the estimated regression model Y2 is approximately {round(x_mle2, 3)} and they are statistically independent, based on visual inspection.") 
```

```python
plt.figure(figsize=(5,5))
plt.scatter(Y1, X, color='blue')
plt.title("Scatter Plot for X and Y1", fontsize=fontsize)
plt.plot(Y1, x1, color='red', linewidth=5)
plt.ylabel("X")
plt.xlabel("Y1")
plt.grid()


plt.figure(figsize=(5,5))
plt.scatter(Y2, X, color='blue')
plt.title("Scatter Plot for X and Y2", fontsize=fontsize)
plt.plot(Y2, x2, color='red', linewidth=5)
plt.ylabel("X")
plt.xlabel("Y2")
plt.grid()

print(f"> The estimate of the coefficient of Y1 in the estimated regression model X is approximately {round(y_mle1, 3)} and they are statistically independent, based on visual inspection.")  
print(f"> The estimate of the coefficient of Y2 in the estimated regression model X is approximately {round(y_mle2, 3)} but they are NOT statistically independent, based on visual inspection.") 
```

## iv). Is the causal direction between X and Y identifiable from their joint distribution in this case? Briefly explain why.
For the first model, Y1 we can say that their direction is **not identifiable** since all of the residuals are independent. However for the second model Y2, we can say that the directions for X and Y2 are **identifiable**, this example is also known as the **Linear Non-Gaussian Model**.
